## 2: Preprocess the data for training and create the LSTM architecture

### 2.1: Import libraries from tensorflow for preprocessing and LSTM architecture

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences 
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
vocab_size = 10000 #vocabulary size
```

```python
temp = X.copy()
temp['title'][0]
```

```python
content = temp.reset_index()
content
```

```python
content['title'][1]
```

### 2.2: Preprocess data

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
complete=[]

for i in range(0,len(content)):
    title = re.sub('[^a-zA-Z]', ' ', str(content['title'][i])) # remove all symbols/numbers 
    title = title.lower() # lowercase
    title = title.split()
    title = [ps.stem(word) for word in title if not word in stopwords.words('english')] # remove stopwords and perform stemming
    title = ' '.join(title)
    complete.append(title)
```

```python
print(len(complete))
complete[:10]
```

### 2.3: One-hot representation

```python
one_hot_encoded = [one_hot(titles, vocab_size) for titles in complete]
```

```python
one_hot_encoded
```

### 2.4 Word embedding representation

```python
sentence_max_length = 25
embedded_complete = pad_sequences(one_hot_encoded, maxlen = sentence_max_length, padding='pre') # padding the one-hot encoded data 
```

```python
embedded_complete
```

```python
embedded_complete[0]
```

### 2.5: Creating the LSTM architecture

```python
embedding_features = 50
model = Sequential()
model.add(Embedding(vocab_size, embedding_features, input_length = sentence_max_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dropout(0.3))
#model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_main = np.array(embedded_complete)
y_main = np.array(y)
```

```python
print(X_main.shape)
X_main
```

```python
print(y_main.shape)
y_main
```

## 3: Training the model 
### 3.1: Splitting the data into train and test set 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_main, y_main, test_size=0.3)
```

### 3.2: Training

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=64)
```

## 4: Results

### 4.4: Function to predict fake or not for a single sentence

```python
def fake_or_not(sentence):
    sent = sentence
    title = re.sub('[^a-zA-Z]', ' ', sent)
    title = title.lower()
    title = title.split()
    title = [ps.stem(word) for word in title if not word in stopwords.words('english')]
    title = ' '.join(title)
    one_hot_sent = [one_hot(title, vocab_size)]
    sent_pad = pad_sequences(one_hot_sent, padding='pre', maxlen=sentence_max_length)
    prediction = int(model.predict_classes(sent_pad))
    if prediction == 1:
        print("Fake!!")
    else:
        print("Not Fake!!")
```
